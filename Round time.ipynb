```python
tweets_df = pd.read_csv('tweets (3).csv',encoding='ISO-8859-1')
tweets_df
```

```python
from datetime import datetime, timedelta

# Function to round time to the nearest hour based on minutes
def round_to_nearest_hour(created_at):
    # Parse the created_at string into a datetime object
    dt = datetime.strptime(created_at, '%m/%d/%Y-%H:%M:%S')
    # Round down if minutes < 30, else round up to the next hour
    if dt.minute < 30:
        rounded = dt.replace(minute=0, second=0)
    else:
        rounded = (dt + timedelta(hours=1)).replace(minute=0, second=0)
    # Format back to the desired string format
    return rounded.strftime('%m/%d/%Y-%H:%M')

# Apply the rounding function and create a new column
tweets_df['rounded_time'] = tweets_df['created_at'].apply(round_to_nearest_hour)

# Create a new table with the relevant columns
rounded_time_df = tweets_df[['created_at', 'rounded_time']]

# Display the new table to the user
import ace_tools as tools; tools.display_dataframe_to_user(name="Rounded Time Table", dataframe=rounded_time_df)
```

```python
# Function to check if a string is a valid datetime in the expected format
def is_valid_datetime(date_str, format='%m/%d/%Y-%H:%M:%S'):
    try:
        datetime.strptime(date_str, format)
        return True
    except ValueError:
        return False

# Filter the dataframe to keep only rows with valid datetime strings
valid_tweets_df = tweets_df[tweets_df['created_at'].apply(is_valid_datetime)]

# Apply the rounding function to valid rows
valid_tweets_df['rounded_time'] = valid_tweets_df['created_at'].apply(round_to_nearest_hour)

# Create a new table with the relevant columns
rounded_time_df = valid_tweets_df[['created_at', 'rounded_time']]
tweets_df['created_at'] = tweets_df['created_at'].astype(str)
valid_tweets_df = tweets_df[tweets_df['created_at'].apply(is_valid_datetime)]

# Apply the rounding function to valid rows
valid_tweets_df['rounded_time'] = valid_tweets_df['created_at'].apply(round_to_nearest_hour)

# Create a new table with the relevant columns
rounded_time_df = valid_tweets_df[['created_at', 'rounded_time']]

rounded_time_df
```
